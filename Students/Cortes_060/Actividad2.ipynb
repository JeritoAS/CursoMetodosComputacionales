2. El potencial Lennard-Jones es un modelo simple que aproxima la interacción entre un par de moléculas neutras. Este está definido como:

$$
V(r) = 4\epsilon \left[ \left( \frac{\sigma}{r} \right)^{12} - \left( \frac{\sigma}{r} \right)^{6} \right]
$$

Donde:
- $ V(r) $ es el potencial en función de la distancia interatómica $ r $.
- $ \epsilon $ es la profundidad del pozo de potencial, es decir, la energía mínima que describe la interacción entre dos partículas.
- $ \sigma $ es la distancia a la cual el potencial es cero, también conocido como el parámetro de distancia.
- $ r $ es la distancia entre las dos partículas.

`````python
5#2.a#

import numpy as np
import matplotlib.pyplot as plt

def potencial_LJ(r, epsilon, sigma):
  potencial = 4*epsilon*((sigma/r)**12 - (sigma/r)**6)
  return potencial

r = float(input(print("Ingrese un valor para r")))
epsilon = float(input(print("Ingrese un valor para epsilon")))
sigma = float(input(print("Ingrese un valor para sigma")))
respuesta = potencial_LJ(r, epsilon, sigma)
print(respuesta)
`````

`````python
 #2.b#

sigma = 1.0
epsilon = 1.0
rc = 2.5*sigma
r = np.linspace(0.9*sigma,rc,500)

potencial = potencial_LJ(r, epsilon, sigma)


plt.plot(r,potencial,label='Potencial LJ')
plt.axvline(x=2**(2/16)*sigma, color='red',linestyle='--',label='$r_m = 2^{2/16} \sigma$')
plt.axhline(y=0, color='black',linestyle='--')
plt.title("Potencial Lennard-Jones en función del radio")
plt.xlabel("Distancia interatómica ($\sigma$)")
plt.ylabel("Potencial $V(r)$")
plt.legend()
plt.grid(True)
plt.show()
`````

`````python
#2.c#

epsilon = 1.0
sigmas = [1.0,1.05,1.1,1.15]
rc = 2.5 * max(sigmas)
r = np.linspace(0.9*min(sigmas),rc,500)

for sigma in sigmas:
  potencial = potencial_LJ(r,epsilon,sigma)
  plt.plot(r,potencial, label=f"$\sigma$ = {sigma}")

plt.axvline(x=2**(2/16)*sigma, color='red',linestyle='--',label='$r_m = 2^{2/16} \sigma$')
plt.axhline(y=0, color='black',linestyle='--')
plt.xlabel(r'Distancia interatómica ($\sigma$)')
plt.ylabel(r'Potencial V(r)')
plt.title('Potencial de Lennard-Jones para diferentes valores de $\sigma$', fontsize=14)
plt.legend()
plt.grid(True)

`````

`````python
df2 = pd.DataFrame(income)
datos_limpios2 = df2.dropna()

datos_limpios2.rename(columns={'Country':'País', 'Total Wealth (billions USD)':'PIB Total (miles de millones de dólares)'}, inplace = True)
`````

`````python
df_merge = pd.merge(datos_limpios, datos_limpios2, on='País')
df_merge
`````

`````python
top10 = df_merge.head(10)
last10 = df_merge.tail(10)

df_combinados = pd.concat([top10, last10])
df_combinados
`````

`````python
plt.scatter(df_combinados['Altura media, niños, 19 años (cm)'], df_combinados['PIB Total (miles de millones de dólares)'], color='b', label='Niños')
plt.scatter(df_combinados['Niñas, 19 años (cm)'], df_combinados['PIB Total (miles de millones de dólares)'], color='r', label='Niñas')
plt.title("Relación entre altura media a los 19 años y PIB")
plt.xlabel("Altura media (cm)")
plt.ylabel("PIB (miles de millones de dólares)")
plt.legend()
plt.grid(True)
plt.show()
`````

Se evidencia una clara tendencia de la altura media a aumentar en países con mayores PIB.
