```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# Ec de movimiento
global g
g = 9.8

def posx(t,  xo, vo, theta):
  return xo+(vo*np.cos(theta))*t

def posy(t,  vo, theta):
  return (vo*np.sin(theta))*t-0.5*g*t**2


def tv(vo, theta):
  tvuelo = 2*vo*np.sin(theta)/g
  return tvuelo
```

```python
vo=2 #m/s
theta = 30 #deg

theta = np.deg2rad(theta)
tvuelo = tv(vo, theta)

t=np.linspace(0, tvuelo, 100)
x = posx(t, 0, vo, theta)
y = posy(t,  vo, theta)
```

```python
fig, ax  = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
X = np.array([x,y])
np.savetxt("parabolico.txt",X.T)
```
