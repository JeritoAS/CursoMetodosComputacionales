# Newton-Rapshon
\begin{equation}
x_{n+1}=x_n - \frac{f(x)}{f'(x)}
\end{equation}

```python
f = lambda x: x*np.exp(x)-1
fp = lambda x: x*np.exp(x) + np.exp(x)
def newton_rapshon(f, fp,x0 , epsilon=1E-9, Nmax=100):
  iter=0
  while((abs(f(x0))>=epsilon) and (iter<=Nmax) ):
    x1 =x0-f(x0)/fp(x0)
    x0=x1
    iter=iter+1
  return x0
```

```python
x0 = newton_rapshon(f, fp, 0.2)
x0
```
